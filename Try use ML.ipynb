```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv('CLEAR_DATASET.csv')
del data['Unnamed: 0']
```

```python
data.head()
```

```python
input = ['40-50', 'Camera']

input = {'price': 45000.0, 'Camera':None, 'memory': None, 
         'battery':None, 'display':None, 'hz':None}
```

```python
models = None

with open('models.pickle', 'rb') as file:
    models = pickle.load(file)
    
```

```python
models
```

```python
X = np.array(data.price).reshape(-1, 1)
cluster_price = models['price'].predict(X)
to_send_data = data
for feature, cond in input.items():
    if cond is not None:
        predict = models[feature].predict([[cond]])
        print(predict)
to_send_data['price_pred'] = cluster_price
to_send_data = to_send_data[to_send_data['price_pred'] == predict[0]]
to_send_data
```

```python
to_send_data = to_send_data.sort_values(by=['price', 'camera'], ascending=False).head(3)
to_send_data = to_send_data[['price', 'smartphone']]

to_send_data
```
